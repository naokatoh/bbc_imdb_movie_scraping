```python
# But some geocodes belong to more than two directors!!
# So make ramdom colored points

def rand_color():
    letters = 'ABCDEF'
    import random
    from random import randint
    this_letter = random.choice(letters)
    random_num1 = randint(0,9)
    random_num2 = randint(0,9)
    hexColor = '#' + this_letter + this_letter + str(random_num1)+ str(random_num1)+ str(random_num2)+ str(random_num2)
    return hexColor
```

```python
df_new['properties.color'] = df_new.apply(lambda x: rand_color(), axis=1)
```

```python
# Also make random sized points
def rand_size():
    import random
    from random import randint
    random_num3 = randint(3,15)
    radius = str(random_num3)
    return radius
```

```python
# "properties": {"article": "<p>director</p>", "radius":7, "color": "#FFFF00", "group_id": 1, "group_name": " ", "headline":"","name": "University Apartment"}

def process_to_geojson(file):
    geo_data = {"type": "FeatureCollection", "features":[]}
    for row in file:
        this_dict = {"type": "Feature", "properties":{}, "geometry": {}}
        for key, value in row.items():
            key_names = key.split('.')
            if key_names[0] == 'geometry':
                this_dict['geometry'][key_names[1]] = value
            if str(key_names[0]) == 'properties':
                this_dict['properties'][key_names[1]] = value
        geo_data['features'].append(this_dict)
    return geo_data
```

```python
geo_format = process_to_geojson(ok_json)
```
